# while-sats 

```python
# Miljöutsläpp

fåglar=8000
antalår=0
print(f"år {antalår}: antal fåglar:{fåglar}st")
while fåglar >=800:
    fåglar *=0.5
    antalår +=1
    print(f"år {antalår}: antal fåglar:{fåglar}st")
```

```python
import matplotlib.pyplot as plt
import numpy as np

b = 1.5e6
surt = 1e7
faktor = 1.5
t = 0

y_b = [b]
while b < surt:
    b = faktor * b
    y_b.append(b)
    t += 1

x_t = np.arange(t + 1)

print(f"det tar {t}h för att mjölken ska surna")
print(f"x = {x_t}")
print(f"y = {y_b}")

plt.plot(x_t, y_b, "*-")
plt.xlabel("timmar(h)")
plt.ylabel("antal bakterier")
plt.title("bakterietillväx")
plt.show()
```
