```python
import numpy as np
import matplotlib.pyplot as plt

t = np.linspace(0,6.028)
def h(t):
    return -4*(t**2) + 24*t + 1

störst = np.max(h(t))
index = np.argmax(h(t))
# np.ma()
# np.argmax()
print(f"maximipunketern: ({t[index]},{störst})")

plt.plot(t,h(t))
plt.plot(t[index], störst, "ro")
plt.title("graf")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
