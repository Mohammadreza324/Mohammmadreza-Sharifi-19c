```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,10)
# y = kx + m
def f(k, x, m):
    y = k*x + m
    return y

print(f(2,x,2))

plt.plot(x,f(2,x,2))
plt.title("räta linje")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
