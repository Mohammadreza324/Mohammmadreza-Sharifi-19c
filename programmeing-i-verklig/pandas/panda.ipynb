```python
import pandas as pd

df = pd.read_csv("sysselsattning.csv", encoding="ISO-8859-1", header=1) 
df_f= df[df["sysselsättning"]=="förvärvsarbetande"]
ja = df_f["2019"].sum()
df_j = df[df["sysselsättning"]=="ej förvärvsarbetande"]
nej = df_j["2019"].sum()
print(f"{ja} personer artbetade och {nej} perosner arbetade inte ")
```

```python
import matplotlib.pyplot as plt
lista = [ja, nej]
namn = "förvärvsarbetade", "ej förvärvsarbetade"
plt.pie(lista,labels=namn)
plt.title("så många förvärvsarbetade och inte förärvsarbetade under 2019")
plt.show()
```

```python
import pandas as pd 

df = pd.read_csv("sysselsattning.csv", encoding="ISO-8859-1", header=1)
df_ej = df[df["sysselsättning"]== "ej förvärvsarbetande"]

df_m = df_ej[df_ej["kön"]== "män"]
man = df_m["2019"].sum()

df_k = df_ej[df_ej["kön"]=="kvinnor"]
kvinna = df_k["2019"].sum()
print(f"mellan ålder 16-44 jobbade {man} män och {kvinna} kvinnor")
```
